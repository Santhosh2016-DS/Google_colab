# **Group By**

```python
import pandas as pd
df_order = pd.DataFrame({
    'order_id':[101, 102, 103, 104, 105, 106],
    'customer':['Arun','Bala', 'Arun', 'Priya', 'Priya', 'Arun'],
    'category':['Electronics', 'Fashion', 'Electronics', 'Appliances', 'Decor', 'Electronics'],
    'amount':[10000, 1500, 5000, 5000, 17000, 12000],
    'order_date':['2026-01-01 10:15',
                  '2026-01-01 10:30',
                  '2026-01-02 09:00',
                  '2026-01-02 18:45',
                  '2026-01-03 11:15',
                  '2026-01-03 19:15']

})
df_order['order_date'] = pd.to_datetime(df_order['order_date'])
df_order
```

**Total Revenue per customer**

```python
total_revenue_per_customer = df_order.groupby('customer')['amount'].sum()
df_final = total_revenue_per_customer.reset_index()
df_final
```

**Average order value per category**

```python
average_order_value_per_category = df_order.groupby('category')['amount'].mean().reset_index()
average_order_value_per_category
```

**Multiple Aggregations**

```python
full_analysis_category = df_order.groupby('category')['amount'].agg(['sum', 'mean', 'min', 'max', 'count', 'std']).reset_index()
full_analysis_category
```

**Daily Revenue**

```python
daily_revenue = df_order.groupby(df_order['order_date'].dt.date)['amount'].sum().reset_index()
daily_revenue
```

```python
hourly_data = df_order.groupby(df_order['order_date'].dt.hour)['amount'].sum().reset_index()
hourly_data
```

**Multiple Columns groupby**

```python
df_order.groupby(['customer', 'category'])['amount'].sum().reset_index()
```

**Daily Sales per  category**

```python
df_order.groupby([df_order['order_date'].dt.date, 'category'])['amount'].sum().reset_index()
```

**Pivot Table**

```python
import pandas as pd
df_order = pd.DataFrame({
    'order_id':[101, 102, 103, 104, 105, 106],
    'customer':['Arun','Bala', 'Arun', 'Priya', 'Priya', 'Arun'],
    'category':['Electronics', 'Fashion', 'Electronics', 'Appliances', 'Decor', 'Electronics'],
    'amount':[10000, 1500, 5000, 5000, 17000, 12000],
    'order_date':['2026-01-01 10:15',
                  '2026-01-01 10:30',
                  '2026-01-02 09:00',
                  '2026-01-02 18:45',
                  '2026-01-03 11:15',
                  '2026-01-03 19:15']

})
df_order['order_date'] = pd.to_datetime(df_order['order_date'])
df_order
```

```python
pd.pivot_table(df_order, index = 'customer', columns = 'category', values = 'amount', aggfunc = 'sum')
```

```python
pd.pivot_table(df_order, index = 'customer', columns = 'category', values = 'amount', aggfunc = 'min')
```

```python
pd.pivot_table(df_order, index = ['customer', 'order_id'], columns = 'category', values = 'amount', aggfunc = 'max')
```

```python
pd.pivot_table(df_order, index = 'customer', columns = 'category', values = 'amount', aggfunc = 'mean')
```

```python
pd.pivot_table(df_order, index = 'category', columns = 'customer', values = 'amount', aggfunc = 'sum')
```
