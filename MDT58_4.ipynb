```python
# MANOVA
independent - input features
dependent - target Features

When to use MANOVA
1. Two or more dependent variables (continuous)
2. One or more independent variables (Categorical)
3. suspecting if two or more dependent variables are correlated
```

```python
Cholestral, BMI, BP, SMOKING, UNHEALTHY FOOD, STRESS - Input Features (Independent Features)

 0.35 * (BP)
 0.15 * (Cholestral)
 0.10 * (BMI)
 0.25 * (SMOKING)
 0.10 * (STRESS)
 0.05 * (UNHEALTHY FOOD)

TARGET (0 | 1)
Death percentage (0.00 -  1.00)- Heart attack Output Feature (Dependent)
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from statsmodels.multivariate.manova import MANOVA
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.rename(columns = {'sepal length (cm)': 'sepal_length', 'sepal width (cm)': 'sepal_width',
                     'petal length (cm)': 'petal_length', 'petal width (cm)': 'petal_width'}, inplace = True)
formula = 'sepal_length + sepal_width + petal_length + petal_width ~ species'
maov = MANOVA.from_formula(formula, data = df)
print(maov.mv_test())
```

```python
Wilks lambda: It is the ratio of within group variance to the total variance in the multivariate model 0.0234

Wilks Lambda (Range Values from 0 to 1)
Considered Groups (G1 (sepal_length), G2 (sepal_width), G3(petal_length), G4(petal_width))

Closer to 0 - Groups which we have considered are very different from other groups
Closer to 1 - Groups which we have considered are not very different (i.e. similar) with other groups

# species 3 types of outcome - setosa, Virginica, Versicolor
Considered Groups (G1 (sepal_length), G2 (sepal_width), G3(petal_length), G4(petal_width))
```
