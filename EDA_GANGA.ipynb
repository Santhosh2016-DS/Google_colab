| **> 60%**                           | Drop Column                                   | Too sparse—difficult to trust or restore.                                      | Imputation not recommended                                                  |

| Column | Type        | % Null | Imputation Method                                  | Reason                                                                                                              |
| ------ | ----------- | ------ | -------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------- |
| City   | Categorical | 20%    | **ML-based prediction (Random Forest Classifier)** | City can depend on Department, State, etc. Predicting missing city using ML improves accuracy over mode imputation. |


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Step 1: Copy the dataset
df_temp = df.copy()

# Step 2: Drop columns not useful for prediction
drop_cols = ['Employee_ID', 'Employee_Name', 'Email', 'Phone', 'Manager_Name', 'City',
             'Date_of_Joining', 'Last_Promotion_Date']  # <-- Drop datetime columns here
df_model = df_temp.drop(columns=drop_cols)

# Step 3: Label encode all categorical columns
le_dict = {}
for col in df_model.select_dtypes(include='category').columns:
    le = LabelEncoder()
    df_model[col] = df_model[col].astype(str)
    df_model[col] = le.fit_transform(df_model[col])
    le_dict[col] = le

# Step 4: Separate data into known and unknown City rows
df_with_city = df[df['City'].notnull()]
df_without_city = df[df['City'].isnull()]

X = df_model.loc[df_with_city.index]
y = df_with_city['City'].astype(str)  # target must be string or int

# Step 5: Encode City
le_city = LabelEncoder()
y_encoded = le_city.fit_transform(y)

# Step 6: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, random_state=42)

# Step 7: Train the model
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# Step 8: Predict missing City values
X_pred = df_model.loc[df_without_city.index]
y_pred = clf.predict(X_pred)
city_filled = le_city.inverse_transform(y_pred)

# Step 9: Fill missing values
df.loc[df['City'].isnull(), 'City'] = city_filled
```

⚙️ What is MICE (Iterative Imputer)?
MICE (Multiple Imputation by Chained Equations) uses a regression model to predict missing values by learning from other columns in sequence.

It’s great for cases where features are interdependent.



✅ Simple Code for Imputing State with MICE
We will:

Convert categorical columns to numeric using Label Encoding.

Apply IterativeImputer (from sklearn) on the encoded data.

Replace the missing values with predicted ones.

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Step 1: Copy your original dataframe
df_temp = df.copy()

# Step 2: Drop pure-text columns (non-usable for imputation)
drop_cols = ['Employee_ID', 'Employee_Name', 'Email_ID']
df_temp = df_temp.drop(columns=drop_cols, errors='ignore')

# Step 3: Convert datetime columns to integers
for col in df_temp.select_dtypes(include='datetime').columns:
    df_temp[col] = df_temp[col].astype('int64')

# Step 4: Encode categorical columns with NaNs using LabelEncoder safely
label_encoders = {}
for col in df_temp.select_dtypes(include='object').columns:
    df_temp[col] = df_temp[col].astype(str)  # for LabelEncoder to work
    le = LabelEncoder()
    df_temp[col] = le.fit_transform(df_temp[col])
    label_encoders[col] = le  # Save for inverse if needed

# Step 5: Use SimpleImputer for Numeric columns
num_imputer = SimpleImputer(strategy='median')
num_cols = df_temp.select_dtypes(include=['int64', 'float64']).columns
df_temp[num_cols] = num_imputer.fit_transform(df_temp[num_cols])

# Step 6: Now the data is clean and numeric — ready for advanced imputation later like MICE
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Step 7: MICE Imputation on 'State' (and related columns)
# Pick 'State' and a few other helpful columns for better prediction
mice_cols = ['State', 'Country', 'Age', 'Salary']  # Choose features correlated to 'State'
mice_cols = [col for col in mice_cols if col in df_temp.columns]

# Create a subset with those columns
df_mice_subset = df_temp[mice_cols]

# Apply MICE
mice_imputer = IterativeImputer(random_state=42)
df_mice_imputed_array = mice_imputer.fit_transform(df_mice_subset)

# Convert back to DataFrame
df_mice_imputed = pd.DataFrame(df_mice_imputed_array, columns=mice_cols)

# Step 8: Replace the imputed 'State' values back into df_temp
df_temp['State'] = df_mice_imputed['State'].round().astype(int)  # round if it was LabelEncoded

```

| **Data Type**             | **Recommended Imputation Techniques**                            | **When to Use**                                                                                                    |
| ------------------------- | ---------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------ |
| **Numeric (Continuous)**  | - Mean<br>- Median<br>- MICE<br>- KNN                            | - Mean: when data is symmetric<br>- Median: when data is skewed<br>- MICE/KNN: when correlated with other features |
| **Numeric (Discrete)**    | - Mode<br>- Median                                               | - Mode: when values are repeated<br>- Median: when treated like ranked data                                        |
| **Categorical (Nominal)** | - Mode<br>- Label Encoding + MICE/KNN                            | - Mode: for few unique categories<br>- MICE/KNN: when influenced by other features                                 |
| **Categorical (Ordinal)** | - Ordinal Encoding + Median<br>- MICE                            | - When order matters and needs to be preserved                                                                     |
| **Datetime**              | - Fill with median/mode date<br>- Domain-specific default values | - When dates are partly missing or a logical default can be inferred                                               |
| **Text / Identifiers**    | - Drop or exclude from imputation                                | - When columns are free-text or unique IDs with no predictive value                                                |

📌 What are Outliers?
Outliers are data points that differ significantly from the majority of the data.
They may occur due to variability in data, data entry errors, or rare events.



🧠 Why are Outliers Important?
They can distort statistical summaries (mean, standard deviation).

They may impact model performance, especially in algorithms sensitive to scale.

They can reveal valuable insights (e.g., fraud detection, rare diseases).



🎯 Why Should You Handle Outliers?
✅ They can affect model accuracy
✅ They may distort mean, variance, correlations
✅ They can indicate special cases like fraud 💳, system failure 💥, or VIP customers 🧑‍💼

| Strategy               | When to Use                                      | Explanation                                                      |
| ---------------------- | ------------------------------------------------ | ---------------------------------------------------------------- |
| 🧹 **Remove**          | ❌ When outlier is clearly an error or noise      | E.g., Age = 999, Salary = -100                                   |
| 🧴 **Cap (Winsorize)** | 🤏 When you want to **limit influence**          | Replace extreme values with percentile caps (e.g., 5th and 95th) |
| 🔄 **Transform**       | 🌀 When outliers stretch scale                   | Use `log`, `sqrt`, `Box-Cox` to reduce impact                    |
| 📥 **Impute**          | 🛠️ When outliers are missing or bad sensor data | Use mean/median or model-based estimates                         |
| 👀 **Keep**            | ✅ When outliers are **genuinely meaningful**     | Keep for anomaly detection, fraud cases, rare customers          |


🧠 Here’s When You Should Check for Outliers:

| ✅ Column Type                                                  | ✔️ Check for Outliers? | 🔍 Reason                                                             |
| -------------------------------------------------------------- | ---------------------- | --------------------------------------------------------------------- |
| **Numeric columns** (e.g., `Salary`, `Age`, `Purchase_Amount`) | ✅ Yes                  | These can have unusually high or low values that distort the analysis |
| **Continuous variables** (not fixed categories)                | ✅ Yes                  | Outliers can heavily influence mean, standard deviation, etc.         |
| **Numeric target variables** in regression                     | ✅ Yes                  | Outliers can affect prediction accuracy and error                     |


🚫 You Can Skip Outlier Check for:

| 🚫 Column Type                                         | ❌ Skip Outlier Check | 📌 Why                                               |
| ------------------------------------------------------ | -------------------- | ---------------------------------------------------- |
| **Categorical columns** (e.g., `Gender`, `Department`) | ✅ Skip               | Outliers don't exist in labels/categories            |
| **ID columns** (`Employee_ID`, `Invoice_No`)           | ✅ Skip               | Each value is unique and not meaningful in magnitude |
| **Binary/Flag columns** (`Yes/No`, `0/1`)              | ✅ Skip               | Only two values — no outliers possible               |
**bold text**

✅ 1. Select numeric columns
We’ll apply this only to numeric columns (e.g., Salary, Age, etc.)

✅ 2. Calculate IQR (Interquartile Range)
IQR = Q3 - Q1
Outliers are any values that fall outside the range:
[Q1 - 1.5 * IQR, Q3 + 1.5 * IQR]

```python
# Select only numeric columns
numeric_cols = df_temp.select_dtypes(include=['int64', 'float64']).columns

# Loop through each numeric column and detect outliers using IQR
for col in numeric_cols:
    Q1 = df_temp[col].quantile(0.25)   # First quartile (25%)
    Q3 = df_temp[col].quantile(0.75)   # Third quartile (75%)
    IQR = Q3 - Q1                      # Interquartile Range

    lower_bound = Q1 - 1.5 * IQR       # Lower limit for outliers
    upper_bound = Q3 + 1.5 * IQR       # Upper limit for outliers

    outliers = df_temp[(df_temp[col] < lower_bound) | (df_temp[col] > upper_bound)]

    print(f"📌 Column: {col} — Found {len(outliers)} outliers")
```

✅ Outlier Summary (Using IQR method)

| Column            | Outliers Found |
| ----------------- | -------------- |
| **Age**           | **215** ⚠️     |
| All other columns | 0 ✅            |


🎯 Interpretation:
✅ Good news: Most columns have no outliers — data looks clean.

⚠️ Concern: Age has 215 outliers, which is significant and needs attention.



💡 What to Do Next for Age?
We have 3 options to handle these 215 outliers:

| Method                               | Description                                                             | Keep Original Distribution? | Recommended When                            |
| ------------------------------------ | ----------------------------------------------------------------------- | --------------------------- | ------------------------------------------- |
| **1. Remove** ❌                      | Simply drop the rows with outlier values.                               | No ❌                        | When outliers are clearly wrong             |
| **2. Cap (Winsorize)** 🧢            | Replace extreme values with boundary values (Q1-1.5*IQR or Q3+1.5*IQR). | Yes ✅                       | When values are extreme but still realistic |
| **3. Treat separately / binning** 🧩 | Convert into categories or handle with flags.                           | Partial                     | When data is skewed                         |


✅ Let’s Apply Capping (Winsorization) for Age — Most Balanced Approach

```python
# Calculate bounds
Q1 = df_temp['Age'].quantile(0.25)
Q3 = df_temp['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Apply capping (Winsorization)
df_temp['Age'] = df_temp['Age'].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)
```

🧠 What is Winsorization?
Winsorization is a technique used to limit extreme values (outliers) in your data without removing them completely. Instead of deleting them 🚫, we replace them with a threshold value (like a percentile).

🧮 How It Works (Step-by-Step)
1️⃣ Identify the percentile thresholds
We choose two cut-off points:

Lower percentile (e.g., 5%)

Upper percentile (e.g., 95%)

📊 Example:
In a list of salaries, the lowest 5% and the highest 5% values are considered too extreme.

2️⃣ Replace the outliers
All values below the 5th percentile ⬇️ are set to the 5th percentile value

All values above the 95th percentile ⬆️ are set to the 95th percentile value

🔧 This pulls in the extremes closer to the center without deleting them.

🎯 Why Use Winsorization?
✅ Keeps your dataset size the same
✅ Controls the impact of extreme values
✅ Works well with mean-based models like linear regression
✅ Safer than removing data altogether



🧪 Analogy with Example
Imagine you're in a classroom 🧑‍🏫

98 students scored between 60 and 90

1 student scored 0 😴

1 student scored 100 🤯

Without Winsorization:

Mean score is heavily affected by 0 and 100

With Winsorization:

Replace 0 with 5th percentile (say 60)

Replace 100 with 95th percentile (say 90)

🎉 Now your dataset is more stable and reliable!



Too high? ⬆️ Clamp it to 95th percentile  
Too low? ⬇️ Clamp it to 5th percentile  
Everything in between? ✅ Leave as is!

🎯 What is Skewness?
Skewness tells us how symmetric the data is around its mean.

⚖️ Skewness ≈ 0 → Data is symmetric (normal-like distribution)

📈 Positive skew → Tail is longer on the right (few high values)

📉 Negative skew → Tail is longer on the left (few low values)



🤔 Why Check Skewness After Outlier Removal?
Outliers can distort the shape of your data distribution.

So after you remove or cap them:

✅ 1. To Verify Distribution Improvement
Outliers cause heavy skewing.

After removal (e.g., using IQR or Winsorization), the data becomes more normally distributed.

We check skewness again to see if data became more balanced.

✅ 2. To Decide Next Steps for Feature Engineering
If skewness is still high (say > ±1), we might still apply:

Log transformation (log(x+1) for right skew)

Box-Cox or Yeo-Johnson transformation

✅ 3. For Better Model Performance
Many ML models (like linear regression) assume normal distribution of data.

Reducing skewness after outlier handling helps:

📉 Reduce prediction errors

📈 Improve accuracy

🧪 Example:
Suppose you had:

Monthly_Income with skewness = 2.8 🧨

You removed outliers using IQR

New skewness = 0.7 ✅
✅ This tells us the distribution is now much closer to normal.

📌 Summary:
Step	Purpose of Skewness Check
📉 Before	Detect skew caused by outliers
🧼 After	Confirm that outlier handling worked and data shape is improved
🛠️ Still skewed?	Apply transformation

```python
import pandas as pd
from scipy.stats import skew

# Step 1: Load your cleaned dataset (null values already handled)
# Example: df is your working DataFrame

# Step 2: Check original skewness
original_skew = df['Age'].skew()
print(f"📊 Original Skewness of Age: {original_skew:.2f}")

# Step 3: Handle outliers using IQR (as an example)
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Remove outliers (or you can cap instead)
df_outlier_removed = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

# Step 5: Check skewness after outlier removal
new_skew = df_outlier_removed['Age'].skew()
print(f"✅ Skewness of Age after outlier removal: {new_skew:.2f}")
```

🔍 Feature Selection: Overview

Feature selection is the process of choosing the most relevant input variables (features) for your predictive model. It improves:

Model accuracy

Training speed

Interpretability

Feature selection can be driven by statistical analysis and visualizations at three levels:

1️⃣ Univariate Analysis (One Variable at a Time)
🧠 What it is:
Examining each feature independently, focusing on its distribution, central tendency, and relationship with the target variable.

✅ Purpose:
Identify irrelevant or low-variance features.

Spot skewed distributions or outliers.

📈 Techniques:

| Feature Type       | Method             | Use Case                                                  |
| ------------------ | ------------------ | --------------------------------------------------------- |
| Numerical          | Variance Threshold | Remove features with low variance                         |
| Categorical        | Chi-Square Test    | Check dependency with categorical target                  |
| Numerical & Target | ANOVA F-test       | Check significance of numerical feature in classification |


📊 Visualizations:
Histogram / KDE Plot (for distributions)

Box Plot (for outliers)

Bar Plot (for categorical count)



2️⃣ Bivariate Analysis (Two Variables: Feature vs Target)
🧠 What it is:
Studying the relationship between each feature and the target variable.

✅ Purpose:
Identify strong individual predictors.

Detect linear/non-linear associations.

📈 Techniques:

| Feature Type vs Target     | Method              | Use Case                        |
| -------------------------- | ------------------- | ------------------------------- |
| Numeric vs Numeric         | Pearson Correlation | Linear relation in regression   |
| Categorical vs Target      | Chi-Square Test     | Dependency for classification   |
| Numeric vs Categorical     | ANOVA / t-test      | Effect of group on a number     |
| Categorical vs Categorical | Cramér’s V          | Measure strength of association |


📊 Visualizations:
Scatter Plot (Numeric vs Numeric)

Box Plot (Numeric vs Category)

Bar Plot (Category vs Target Mean)

Heatmap of Correlation Matrix

3️⃣ Multivariate Analysis (Multiple Features Interacting)
🧠 What it is:
Analyzing more than two features together to understand inter-feature relationships or joint effects on the target.

✅ Purpose:
Detect multicollinearity

Reveal interaction effects

Select features in the presence of redundancy

🧠 Visualization & Feature Selection Recommendation Matrix

| Analysis Type | Key Use Case                           | Recommended Visuals               | Recommendation                                         |
| ------------- | -------------------------------------- | --------------------------------- | ------------------------------------------------------ |
| Univariate    | Understand single feature distribution | Histogram, Box Plot               | Drop constant or low-variance features                 |
| Bivariate     | Assess feature-target relationship     | Correlation Heatmap, Bar/Box Plot | Keep high-impact features                              |
| Multivariate  | Detect redundancy or synergy           | Heatmap, Pairplot, VIF Chart      | Remove highly correlated features or reduce dimensions |


✅ Final Recommendation Strategy
Start with Univariate – Remove features with low variance or irrelevant distributions.

Proceed with Bivariate – Check which features have significant relationships with the target.

Apply Multivariate Analysis – Avoid multicollinearity and refine selection using model-based techniques.

🎯 What Is Data Imbalance?
Data imbalance occurs when one class (or category) in the target variable has significantly more instances than others.

For example:

Loan Approved: Yes (9800) vs No (200)

Attrition: No (85%) vs Yes (15%)



🧾 How to Summarize Imbalance?
Check Class Distribution:

```python
# For classification target column like 'Attrition'
df['Attrition'].value_counts(normalize=True) * 100  # shows % of each class
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Attrition')
plt.title("Class Distribution")
plt.show()
```

```python
# Example for binary classification
ratio = df['Attrition'].value_counts()
imbalance_ratio = ratio.max() / ratio.min()
print(f"Imbalance Ratio: {imbalance_ratio:.2f}")
```

| **Imbalance Ratio (IR)** | **Interpretation**                       | **Severity Level** | **Action Required?**                     |
| ------------------------ | ---------------------------------------- | ------------------ | ---------------------------------------- |
| **1.0 – 1.5**            | Nearly balanced classes                  | 🔹 Low             | Usually no action needed                 |
| **1.6 – 2.5**            | Slight imbalance                         | ⚠️ Moderate        | Monitor, minor adjustments               |
| **2.6 – 5.0**            | Noticeable imbalance                     | 🔸 High            | Consider handling imbalance              |
| **5.1 – 10.0**           | Significant imbalance                    | 🔺 Very High       | Imbalance handling recommended           |
| **> 10.0**               | Extremely imbalanced (e.g., 90:10, 95:5) | 🚨 Critical        | Must handle (e.g., SMOTE, class weights) |


🧰 What To Do If Data Is Imbalanced?

| **Technique**         | **What It Does**                              | **When to Use**                    |
| --------------------- | --------------------------------------------- | ---------------------------------- |
| **Undersampling**     | Removes samples from majority class           | When data is large                 |
| **Oversampling**      | Replicates minority class samples             | When minority class is small       |
| **SMOTE**             | Creates synthetic samples for minority class  | For balanced representation        |
| **Class Weights**     | Modifies model training to penalize imbalance | Use in LogisticRegression, XGBoost |
| **Anomaly Detection** | Treat minority class as anomaly               | When class imbalance is extreme    |
