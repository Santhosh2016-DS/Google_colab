```python
# Calculating a Product's Discounted Price (E-commerce/Retail)
def calculate_discounted_price(original_price, discount_percent):
    """Calculates the final price after applying a discount."""
    # Convert percentage to a decimal
    discount_factor = discount_percent / 100

    # Calculate the discount amount
    discount_amount = original_price * discount_factor

    # Calculate and return the final price
    final_price = original_price - discount_amount

    return final_price

# Example Usage
original_cost = 150.00
sale_discount = 20
final_cost = calculate_discounted_price(original_cost, sale_discount)

print(f'Calculation of Discounted Price')
print(f'Original Price: {original_cost}')
print(f"Discount: {sale_discount}%")
print(f"Final Price: {final_cost:.2f}")
```

```python
# Validating User's Password
def is_password_strong(password):
    """
    Checks if a password meets minimum strength requirements:
    - At least 8 characters long
    - Contains at least one uppercase letter
    - Contains at least one digit
    """
    # Rule 1: Minimum length check
    print('length of password = ', password, len(password))
    if len(password) < 8:

        return False

    # Rule 2: Uppercase check
    print('length of password = ', password, len(password))
    if not any(char.isupper() for char in password):
        return False

    # Rule 3: Digit check
    if not any(char.isdigit() for char in password):
        return False

    # If all rules pass
    return True

# Example Usage
password_a = "weakpass"
password_b = "SecureP@ss1"

print(f"'{password_a}' is strong: {is_password_strong(password_a)}")
print(f"'{password_b}' is strong: {is_password_strong(password_b)}")
```

# Functions as Arguments

```python
# Function as Arguments
def debit(balance, amount_transfered):
  return balance - amount_transfered

def credit(balance, amount_transfered):
  return balance + amount_transfered

def intermediate_function(function_name, X, Y, e_address):
  current_balance = function_name(X, Y)
  print('Inside intermediate_function', X, Y, e_address)
  return current_balance

account_balance, amount_to_be_transfered, email = 1000, 200, 'a@gmail.com'
final_balance = intermediate_function(debit, account_balance, amount_to_be_transfered, email)
print(f'final balance in account = {final_balance}')
```

```python
# A function to extract the age from a user dictionary
def get_age(user_dict):
    return user_dict['age']

# A function to extract the name from a user dictionary
def get_name_lower(user_dict):
    return user_dict['name'].lower()

# The built-in sorted function is a higher-order function.
# We pass our custom function (get_age) as the 'key' argument.
# sorted() will call get_age() for each item to determine the sorting order.

users = [
    {'name': 'John', 'age': 30, 'city': 'New York'},
    {'name': 'Jane', 'age': 25, 'city': 'Los Angeles'},
    {'name': 'Bob', 'age': 35, 'city': 'Chicago'}
]
sorted_by_age = sorted(users, key=get_age)

# We can reuse the same sorted() function but pass a different key function
# to sort by name (case-insensitive).
sorted_by_name = sorted(users, key=get_name_lower)

print("Sorted by Age:")
for user in sorted_by_age:
    print(user)

print("\nSorted by Name:")
for user in sorted_by_name:
    print(user)
```

```python
# Example algorithms - binary search, Euclids algorithm
# Binary Search
def binary_search_algorithm(sample_variable, target_value):
  low_index = 0
  high_index = len(sample_variable) - 1
  while low_index <= high_index:
    middle_index = (low_index + high_index)//2
    print('Low Index = ', low_index)
    print('Middle Index = ', middle_index)
    print('High Index = ', high_index)
    print('********')
    middle_value = sample_variable[middle_index]
    if middle_value == target_value:
      return "Data found in the list"
    elif middle_value < target_value:
      low_index = middle_index + 1
    else:
      high_index = middle_index - 1
  return "Data not found in the list"

sample_list = [2,5,8,12,16,23,38,56,72,91]
print(sample_list, type(sample_list))
tv = 1568 # target
binary_search_algorithm(sample_list, tv)
```

```python
# Euclids algorithm (GCD Algorithm) Greatest Common Divisor
# Divisors of 5 = 1, 5
# Divisors of 10 = 1, 2, 5, 10
# Common Divisors = 1, 5
# Greatest Common Divisors = 5

def euclid_algo(X, Y):
  while Y:
    X, Y  = Y, X % Y
    print('Inside While Loop')
    print('X = ', X)
    print('Y = ', Y)
    print('********')
  return X

value_1, value_2 = 1920, 1080
gcd = euclid_algo(value_1, value_2)
print(gcd)
```

```python
# Peek
def peeking_stack(current_list):
  peeked_value = len(current_list)-1 # -1 or len(current_list)-1
  return peeked_value

# push
def pushing_stack(current_list, new_value):
  current_list.append(new_value)
  return current_list

# pop
def Popping_stack(current_list):
  maximum_index = len(current_list) - 1
  current_list.pop(maximum_index)
  return current_list

# length Checking
def length_checking(sample_list):
  return len(sample_list)

sample_lister = [11, 12, 13, 14, 15]
print('sample_lister = ', sample_lister)
current_length = length_checking(sample_lister)
print('current_length = ', current_length)

#Peeking Operation
peeked_value = peeking_stack(sample_lister)
print('peeked_value = ', sample_lister[peeked_value])

#Pushing Operation
current_list = pushing_stack(sample_lister, 16)
print('current_list = ', current_list)
current_length = length_checking(current_list)
print('current_length = ', current_length)

#Popping Operation
final_result = Popping_stack(current_list)
print('final_result = ', final_result)
current_length = length_checking(current_list)
print('current_length = ', current_length)
```

```python
# Binary Tree (Oops)
class Node:
  # Initialization
  def __init__(self, V):
    self.value = V
    self.left = None
    self.right = None

  def traversing_pre_order(self):
    print('Preorder Root Value = ', self.value)
    if self.left:
      self.left.traversing_pre_order()
    if self.right:
      self.right.traversing_pre_order()

  def traversing_in_order(self):
    if self.left:
      self.left.traversing_in_order()
    print('In order Root Value = ', self.value)
    if self.right:
      self.right.traversing_in_order()

  def traversing_post_order(self):
    if self.left:
      self.left.traversing_post_order()
    if self.right:
      self.right.traversing_post_order()
    print('Post order Root Value = ', self.value)

root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
#Operations
root.traversing_pre_order()
print('Pre order ********')
root.traversing_in_order()
print(' In order ********')
root.traversing_post_order()
print('Post Order********')
```

```python
# Heap (Min Heap and Max Heap)
# Heap is a spcialized binary tree based on data structure that satisfies the heap property (Minimum or Maximum)
# It is mainly used in priority queues

# Min Heap
import heapq
class MinHeap:
    def __init__(self):
        self.heap = []

    def push(self, value):
      print('The value pushed inside the heap tree structure = ', value)
      print('The Before pushing heap tree structure looks like = ', self.heap)
      heapq.heappush(self.heap, value)
      print('The after pushing heap tree structure looks like = ', self.heap)
      print('***************')

    def pop(self):
      if len(self.heap)==0:
        return self.heap
      else:
        value_removed = heapq.heappop(self.heap)
        print('The value popped from the heap tree structure = ', value_removed)
        print('current tree structure = ', self.heap)

    def peek(self):
      if len(self.heap)==0:
        return 'Heap is Empty'
      else:
        value = self.heap[0]
        print('The value peeked from the heap tree structure = ', value)

    def length_checking(self):
      if len(self.heap)==0:
        return 'Heap is Empty'
      else:
        return 'Heap has values'

min_heap = MinHeap()
min_heap.push(77)
min_heap.push(30)
min_heap.push(23)
min_heap.push(31)
min_heap.push(17)
min_heap.push(99)
min_heap.push(99)
min_heap.push(44)
min_heap.push(3)
min_heap.push(0)
min_heap.peek()
min_heap.length_checking()
min_heap.pop()
```

```python
import sys
# Generators vs Lists (memory efficient)
def generate_numbers(n):
    for i in range(n):
        yield i * i

# Generator - memory efficient
gen = generate_numbers(1000000)
print(f"Generator size: {sys.getsizeof(gen)} bytes")

# List - memory intensive
lst = [i * i for i in range(1000000)]
print(f"List size: {sys.getsizeof(lst)} bytes")
```
