**Comparison**

```python
# Histogram
# When to use: Distribution of Continuous Data, Identifying Skewness, Identifying Outliers
# Example: Median income, Test scores, Class age groups, Sales
```

```python
# Box Plot
# When to use: Continuous data (Multiple groups), To detect Outliers, To Identify the spread of the data, To calculate the IQR
# Examples: Salary Distribution by department, 1 Class Exam scores with various subjects
```

```python
# Pie Chart
# When to use: Simple Categorical Values, Dominant categorical propositions (<10 or <15 unique values of categorical proportions)
# Example: UPI Market Shares (Phonepe(65), Gpay(20), paytm(10), paypal(5))
# Limitations: If there is more categories then it is hard to interpret
```

```python
# Bar Plot
# When to use: Categorical Data Categories(Gender: Male, Female, Prefer Not to Say), Displaying both absoulute and relative comparisons
# Example: 10 schools average marks, Region wise sales
```

```python
# Donut Chart (parameter = wedgeprops)
# When to use: Same as pie chart but more good looking
# Examples: Same as pie chart
```

```python
# Stacked Bar Chart
# When to use: Multiple segments in the categories, contribution of sub categories
# Example: Sales of different products over years
```

```python
# Stacked Area Plot
# When to use: used in Time series, Cummulative Value Charting
# Example: you are staying in a house and there is electricity consumption.
# Calculation of total energy consumption over the time period you stayed in that house over years
```

```python
# Relative Stacked Bar Plot (parameter = bottom)
# When to use: used in Time series, Cummulative Value Charting but in a independent values
# Example: Sales of different products over years
```

```python
# Scatter Plots
# When to use: Relationship between two continuous variables, to identify outliers
# Example: Relationship between marketing expense and sales
```

```python
# Line Plot
# When to use: Time series
# Example: Monthly sales over time
```

**Outlier detection and Correction**

```python

IQR Based Outlier Detection
IQR(50%) = Q3(75%) - Q1(25%)
Lower_Limit = Q1 - 1.5 * IQR
Upper_Limit = Q3 + 1.5 * IQR

value < lower_limit = lower bound outlier
value > upper_limit = upper bound outlier
```

# **Bivariate Analysis: Scatter plots and correlation analysis**

```python
import pandas as pd
data = {'hours_studied': [2,4,6,8,10],
        'scores': [40, 55, 70, 85,92]}
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['hours_studied'], df['scores'])
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Hours studied vs Scores')
plt.show()
```

```python
df.corr()
```

```python
INTERPRETATION
Scatter Plot Analysis
1. Upward trend = Positive Correlation
2. More study hours = Higher the marks

Correlation Analysis
1. Hours studied to scores we are getting 0.992947 * 100 ==> 99.29% of very strong possitive correlation
```
